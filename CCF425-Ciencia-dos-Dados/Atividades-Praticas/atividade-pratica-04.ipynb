```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
pd.options.display.float_format = "{:.2f}".format
```

Hipotese Nula: O ia171 tem um tempo de atraso menor do que o ie666

Hipotese Alternativa: O 1e666 tem um tempo de atraso menor do que o ia171

```python
ia171 = pd.read_csv('P04-dados/atraso_ia171.csv', index_col=False,header=None,squeeze=True)
ie666 = pd.read_csv('P04-dados/atraso_ie666.csv', index_col=False,header=None,squeeze=True)
```

```python
ia171.describe()
```

```python
ie666.describe()
```

```python
ax = sns.displot(ia171)
plt.show()
```

```python
ax = sns.displot(ie666)
plt.show()
```

```python
maximum = ie666.mean() + (2 * ie666.std())
minimum = ia171.mean() - (2 * ia171.std())

print(maximum, minimum)
```

```python
ia171_mean = np.mean(ia171)
ie666_mean = np.mean(ie666)
print("ia171 mean value:",ia171_mean)
print("ie666 mean value:",ie666_mean)
ia171_std = np.std(ia171)
ie666_std = np.std(ie666)
print("ia171 std value:",ia171_std)
print("ie666 std value:",ie666_std)
ttest,pval = ttest_ind(ia171,ie666)
print("p-value",pval)
if pval <0.05:
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")
```

Hipotese Nula: O combinar tem um tempo de atraso menor do que o bitcomb

Hipotese Alternativa: O bitcomb tem um tempo de atraso menor do que o combinar



```python
bitcomb = pd.read_csv('P04-dados/tempo_bitcomb.csv', index_col=False,header=None,squeeze=True)
combinar= pd.read_csv('P04-dados/tempo_combinar.csv', index_col=False,header=None,squeeze=True)
```

```python
bitcomb.describe()
```

```python
combinar.describe()
```

```python
ax = sns.displot(bitcomb)
plt.show()
```

```python
ax = sns.displot(combinar)
plt.show()
```

```python
maximum = bitcomb.mean() + (2 * bitcomb.std())
minimum = bitcomb.mean() - (2 * bitcomb.std())

print(maximum, minimum)
```

```python
minimum = combinar.mean() - (2 * combinar.std())
maximum = combinar.mean() + (2 * combinar.std())

print(maximum, minimum)
```

```python
bitcomb_mean = np.mean(bitcomb)
combinar_mean = np.mean(combinar)
print("bitcomb mean value:",bitcomb_mean)
print("combinar mean value:",combinar_mean)
bitcomb_std = np.std(bitcomb)
combinar_std = np.std(combinar)
print("bitcomb std value:",bitcomb_std)
print("combinar std value:",combinar_std)
ttest,pval = ttest_ind(combinar,bitcomb)
print("p-value",pval)
if pval <0.05:
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")
```
