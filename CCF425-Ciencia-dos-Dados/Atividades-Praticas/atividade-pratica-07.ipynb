```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = [['Leite','Cebola','Nutela','Feijão','Ovo','Iogurte'],
['Arroz','Cebola','Nutela','Feijão','Ovo','Iogurte'],
['Leite','Maça','Feijão','Ovo'],
['Leite','Chocolate','Pipoca','Feijão','Iogurte'],
['Pipoca','Cebola','Feijão','Sorvete','Ovo']]
```

```python
# Converte para o formato do Apriori, usando o TransactionEncoder

transactions_oht = TransactionEncoder()
transactions = transactions_oht.fit(dataset).transform(dataset)
dataframe = pd.DataFrame(transactions, columns=transactions_oht.columns_)
dataframe
```

```python
# Calcula os conjuntos de itens frequentes, com suporte mínimo de 0.6

frequent_itemsets = apriori(dataframe, min_support=0.6, use_colnames=True)
frequent_itemsets.sort_values(by="support", ascending=False)
```

```python
# Filtro por 'confidence', pode ser feito por outros campos também
regras = association_rules(frequent_itemsets,metric="confidence", min_threshold=0.1)
regras
```

```python
regras[regras['confidence'] > 0.75]
```

```python
df = pd.read_excel('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx',index_col=False, squeeze=True)
```

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis = 0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[df.Quantity > 0]
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
franca = df[df['Country'] == 'France']
df_grouped = franca.groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
cesta_franca = df_grouped.unstack(fill_value=0)
cesta_franca = cesta_franca.reset_index().set_index('InvoiceNo')
```

```python
def encode_units(x):
    if x <=0:
        return 0
    if x >= 1:
        return 1
cesta_franca = cesta_franca.applymap(encode_units)
cesta_franca.drop('POSTAGE', inplace=True, axis=1)
```

# Atividade

## a) Qual a regra com maior suporte?

```python
frequent_itemsets = apriori(cesta_franca, min_support=0.03, use_colnames=True)
support = association_rules(frequent_itemsets, metric="support", min_threshold=0.05)
support.sort_values(by='support').tail(2)
```

## b) Qual a regra com maior confiança?

```python
confidence = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.05)
confidence.sort_values(by='confidence').tail(2)
```

## c) Qual a regra com maior lift?

```python
lift = association_rules(frequent_itemsets, metric="lift", min_threshold=25)
lift.sort_values(by='lift').tail(2)
```

## d) Quantas regras possuem suporte maior ou igual a 0.02 e confiança maior ou igual a 0.8?

```python
support = association_rules(frequent_itemsets, metric="support", min_threshold=0.02)
len(support[support['confidence'] >= 0.8])
```

## e) Quantas regras possuem lift maior ou igual a 4 e confiança maior ou igual a 0.8?

```python
len(support[(support['lift'] >= 4) & (support['confidence'] >= 0.8)])
```
